```python
import pandas as pd
import numpy as np
import requests
import urllib.request
from sklearn.metrics import precision_recall_curve
# from lxml import html
# from pprint import pprint
# from pymongo import MongoClient
```

Загружаем тестовые данные

```python
path = 'c://users//andrei//downloads//lection9//'
X_test = pd.read_csv(path + 'x_test.csv', index_col=['RowNumber'])
y_test = pd.read_csv(path + 'y_test.csv')
X_test.head()
```

Функция для взаимодействия с сервером.

```python
data = (714, 'Spain', 'Female', 33, 10, 103121.33, 2, 1, 1, 49672.01)
def send_json(info_array):
    
    body = {
        'CreditScore': info_array[0],
        'Geography': info_array[1],
        'Gender': info_array[2],
        'Age': info_array[3],
        'Tenure': info_array[4],
        'Balance': info_array[5],
        'NumOfProducts': info_array[6],
        'HasCrCard': info_array[7],
        'IsActiveMember':info_array[8],
        'EstimatedSalary':info_array[9],
    }
    print(body)
    myurl = 'http://127.0.0.1:5000' + '/predict'
    headers = {'content-type': 'application/json; charset=utf-8'}
    response = requests.post(myurl, json=body, headers=headers)
    return response.json()['Prediction']
```

```python
response = X_test.apply(lambda x: send_json(x), axis=1)
response.values
```

Проверим полученные результаты.

```python
precision, recall, thresholds = precision_recall_curve(y_test['Exited'], response.values)
fscore = (2 * precision * recall) / (precision + recall)
# locate the index of the largest f score
ix = np.argmax(fscore)
print('Best Threshold=%f, F-Score=%.3f, Precision=%.3f, Recall=%.3f' % (thresholds[ix], 
                                                                                       fscore[ix],
                                                                                       precision[ix],
                                                                                       recall[ix]))
```

##### Результаты совпадают!
